```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_path = 'D:/Heart Desease CSV/Heart Disease data/Heart Disease data.csv'
heart_data = pd.read_csv(data_path)

# Display basic information about the dataset
heart_data.info()
print(heart_data.head())
```

```python
# Set visualization style
sns.set(style="whitegrid")

# 1. Age Distribution Plot
plt.figure(figsize=(10, 6))
sns.histplot(heart_data['age'], bins=20, kde=True, color='skyblue')
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Heart Disease Prevalence by Gender
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', hue='target', data=heart_data, palette='coolwarm')
plt.title("Heart Disease Prevalence by Gender")
plt.xlabel("Gender (0 = Female, 1 = Male)")
plt.ylabel("Count")
plt.legend(title="Heart Disease", loc="upper right", labels=["No Disease", "Disease"])
plt.show()

```

```python
# 3. Cholesterol Levels by Heart Disease Status
plt.figure(figsize=(10, 6))
sns.histplot(data=heart_data, x='chol', hue='target', kde=True, palette='viridis', bins=20)
plt.title("Cholesterol Levels by Heart Disease Status")
plt.xlabel("Cholesterol Level")
plt.ylabel("Frequency")
plt.legend(title="Heart Disease", labels=["No Disease", "Disease"])
plt.show()
```
