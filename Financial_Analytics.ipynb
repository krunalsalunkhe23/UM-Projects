```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'D:/Datset/Financial Analytics data.csv' 
data = pd.read_csv(file_path)

# Drop the unnecessary column and rename columns for easier analysis
data_cleaned = data.drop(columns=['Unnamed: 4'])
data_cleaned.columns = ['S.No', 'Company Name', 'Market Cap (Crore)', 'Quarterly Sales (Crore)']
```

```python
# Basic statistics to understand the data distribution
print(data_cleaned.describe())
```

```python
# Top 10 companies by Market Cap
top_market_cap = data_cleaned.nlargest(10, 'Market Cap (Crore)')[['Company Name', 'Market Cap (Crore)', 'Quarterly Sales (Crore)']]
print("Top 10 Companies by Market Cap:")
print(top_market_cap)
```

```python
# Top 10 companies by Quarterly Sales
top_quarterly_sales = data_cleaned.nlargest(10, 'Quarterly Sales (Crore)')[['Company Name', 'Market Cap (Crore)', 'Quarterly Sales (Crore)']]
print("Top 10 Companies by Quarterly Sales:")
print(top_quarterly_sales)
```

```python
# Plotting Market Cap vs. Quarterly Sales to visualize any relationship
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data_cleaned, x='Quarterly Sales (Crore)', y='Market Cap (Crore)')
plt.title("Relationship between Market Cap and Quarterly Sales")
plt.xlabel("Quarterly Sales (Crore)")
plt.ylabel("Market Cap (Crore)")
plt.grid(True)
plt.show()
```
